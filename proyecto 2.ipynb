```python
%matplotlib inline

import matplotlib.pylab as plt
import numpy as np
from sympy import *
import pandas as pd
```

```python

N = 10000
f = lambda x:.0004*x
a,b = 0,500000
Altura = 200

x = np.random.uniform(a,b,N)
y = np.random.uniform(a,Altura,N)
X = np.arange(a,b,.01)
fx = f(x)

plt.plot(x[y>fx],y[y>fx],"go")
plt.plot(x[y<fx],y[y<fx],"ro")
plt.plot(X,f(X))

N_0 = len(x[y>fx])
Base = b-a
yest = N_0/N*(Base*Altura)
print (yest)
```
